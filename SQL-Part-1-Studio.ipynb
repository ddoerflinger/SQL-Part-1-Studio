Question 2: Write a query to find the least popular book.

```sql
-- return least popular book
SELECT TOP 100 book_id, authors, title AS 'Book Title', average_rating FROM BooksDB.dbo.books
ORDER BY average_rating;

-- One Night at the Call Center
```

Question 3: Which tag is the most popular?

```sql
-- return most popular tag
SELECT TOP 10 * FROM BooksDB.dbo.book_tags ORDER BY count DESC;

-- The most popular tag is 30574
```

Question 4: What is the name of the most popular tag?

```sql
-- return name of most popular tag
SELECT * FROM BooksDB.dbo.tags WHERE tag_id = 30574;

-- The name of the most popular tag is to-read
```

Question 5: How many books where released in the first decade of 2000?

```sql
-- return number of books published in the first decade of 2000
SELECT COUNT(*) AS 'Number of Books Released 2000-2010' FROM BooksDB.dbo.books WHERE original_publication_year BETWEEN 2000 and 2010

-- There were 3594 books released between 2000 and 2010.
```

Question 6: How many book titles contain the word Happy?

```sql
-- return number of books that contain the word Happy
SELECT COUNT(*) AS 'Number of Book Titles Containing Happy' FROM BooksDB.dbo.books WHERE title LIKE '%happy%';

-- 13 book titles contain the word happy.
```

Question 9: Write a query that returns the number of Authors whose first name is between rock and roll.

```sql
-- return number of authors whose first name is between rock and roll
SELECT COUNT(authors) AS 'Number of Authors Between Rock and Roll' FROM BooksDB.dbo.books WHERE authors BETWEEN 'rock' AND 'roll';
```
